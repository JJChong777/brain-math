## Prepare Candidate Pools with All Demographics and T2 Task Status

### Subtask:
Create a 'T2_participated' boolean column in `participants_df` and merge this DataFrame with the previously identified `math_difficulty_candidates` and `control_group_candidates` to include demographic information ('sex', 'age_ses-T1', 'handedness', 'T2_participated') and all relevant T2 task status columns.


**Reasoning**:
I will create the 'T2_participated' column, identify and select all relevant demographic and T2 task columns from `participants_df`, and then merge this information with `math_difficulty_candidates` and `control_group_candidates`.



```python
participants_df['T2_participated'] = participants_df['age_ses-T2'].notna()

t2_task_columns = [col for col in participants_df.columns if col.startswith('ses-T2_task-')]
demographics_and_T2_tasks_df = participants_df[['participant_id', 'sex', 'age_ses-T1', 'handedness', 'T2_participated'] + t2_task_columns]

math_difficulty_candidates_demographics_t2 = pd.merge(
    math_difficulty_candidates,
    demographics_and_T2_tasks_df,
    on='participant_id',
    how='inner'
)

control_group_candidates_demographics_t2 = pd.merge(
    control_group_candidates,
    demographics_and_T2_tasks_df,
    on='participant_id',
    how='inner'
)

print("Math Difficulty Candidates with Demographics and T2 Tasks (first 5 rows):")
print(math_difficulty_candidates_demographics_t2.head())
print("\nControl Group Candidates with Demographics and T2 Tasks (first 5 rows):")
print(control_group_candidates_demographics_t2.head())
```

**Reasoning**:
I need to filter both the `math_difficulty_candidates_demographics_t2` and `control_group_candidates_demographics_t2` DataFrames to include only participants who fully participated in and completed *all* specified T2 tasks. This requires ensuring 'T2_participated' is `True` and all 'ses-T2_task-*' columns indicate 'yes'.



```python
t2_task_columns = [col for col in participants_df.columns if col.startswith('ses-T2_task-')]

# Filter math difficulty candidates for T2 participation and all 'yes' in T2 tasks
math_difficulty_candidates_T2_full = math_difficulty_candidates_demographics_t2[
    (math_difficulty_candidates_demographics_t2['T2_participated'] == True) &
    (math_difficulty_candidates_demographics_t2[t2_task_columns] == 'yes').all(axis=1)
]

# Filter control group candidates for T2 participation and all 'yes' in T2 tasks
control_group_candidates_T2_full = control_group_candidates_demographics_t2[
    (control_group_candidates_demographics_t2['T2_participated'] == True) &
    (control_group_candidates_demographics_t2[t2_task_columns] == 'yes').all(axis=1)
]

print("Math Difficulty Candidates after full T2 filter (first 5 rows):")
print(math_difficulty_candidates_T2_full[['participant_id', 'struggle_math', 'WJ-III_MathFluency_StS', 'sex', 'age_ses-T1', 'handedness', 'T2_participated'] + t2_task_columns].head())

print("\nControl Group Candidates after full T2 filter (first 5 rows):")
print(control_group_candidates_T2_full[['participant_id', 'struggle_math', 'WJ-III_MathFluency_StS', 'sex', 'age_ses-T1', 'handedness', 'T2_participated'] + t2_task_columns].head())
```

**Reasoning**:
Now that both the Math Learning Difficulty and Control group candidates have been thoroughly filtered to include only those with full T2 participation, the next step is to implement the group balancing strategy. This involves iterating through the math difficulty candidates, finding the best control match based on sex, handedness, and closest age, and then collecting these matched pairs into final balanced groups.



```python
max_group_size = 15

matched_math_rows = []
matched_control_rows = []

# Create copies to allow removal of selected participants without modifying the original filtered DataFrames
math_candidates_temp = math_difficulty_candidates_T2_full.copy()
control_candidates_temp = control_group_candidates_T2_full.copy()

# Define the columns that should be included in the final balanced groups
columns_to_select = ['participant_id', 'struggle_math', 'WJ-III_MathFluency_StS', 'sex', 'age_ses-T1', 'handedness', 'T2_participated'] + t2_task_columns

for index_math, math_participant in math_candidates_temp.iterrows():
    if len(matched_math_rows) >= max_group_size: # Stop if group is full
        break

    best_match_control = None
    min_age_diff = float('inf')

    # Filter control candidates for exact matches on categorical variables (sex, handedness, T2_participated)
    # T2_participated is already True for both candidate pools, but keeping it in for consistency
    filtered_controls = control_candidates_temp[
        (control_candidates_temp['sex'] == math_participant['sex']) &
        (control_candidates_temp['handedness'] == math_participant['handedness']) &
        (control_candidates_temp['T2_participated'] == math_participant['T2_participated'])
    ]

    # If exact matches found on categorical variables, find the one with closest age
    if not filtered_controls.empty:
        for index_control, control_participant in filtered_controls.iterrows():
            age_diff = abs(math_participant['age_ses-T1'] - control_participant['age_ses-T1'])
            if age_diff < min_age_diff:
                min_age_diff = age_diff
                best_match_control = control_participant

    if best_match_control is not None:
        # Add participants to lists
        matched_math_rows.append(math_participant[columns_to_select].to_dict())
        matched_control_rows.append(best_match_control[columns_to_select].to_dict())

        # Remove selected participants from temporary pools by their original index
        math_candidates_temp = math_candidates_temp.drop(index_math)
        control_candidates_temp = control_candidates_temp.drop(best_match_control.name)

# Convert lists of dictionaries to DataFrames
balanced_math_group_final = pd.DataFrame(matched_math_rows, columns=columns_to_select)
balanced_control_group_final = pd.DataFrame(matched_control_rows, columns=columns_to_select)

print(f"Balanced Math Learning Difficulty Group size: {len(balanced_math_group_final)}")
print(f"Balanced Control Group size: {len(balanced_control_group_final)}")
print("Balanced Math Learning Difficulty Group (first 5 rows):\n", balanced_math_group_final.head())
print("Balanced Control Group (first 5 rows):\n", balanced_control_group_final.head())
```
