```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, f_classif

def calc_show_fstat_score(X_avg, y_avg):
    # Calculate F-statistic scores for all 79,800 features
    selector = SelectKBest(score_func=f_classif, k='all')
    selector.fit(X_avg, y_avg)
    scores = selector.scores_

    # 3. Sort scores in descending order
    sorted_scores = np.sort(scores)[::-1]

    # 4. Create the Plot
    plt.figure(figsize=(10, 6))
    plt.plot(sorted_scores[:1000], color='blue', linewidth=2) # Look at top 1000
    plt.axvline(x=100, color='red', linestyle='--', label='Proposed Cut-off (k=100)')

    plt.title('Univariate Feature Scores (Schaefer 400 Atlas)', fontsize=14)
    plt.xlabel('Feature Rank (Top 1000)', fontsize=12)
    plt.ylabel('F-Statistic Score', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
```

# Verify All Data

```python
X_avg, y_avg = get_all_data()
calc_show_fstat_score(X_avg, y_avg)
```

# Verify Mult Data

```python
X_avg_m, y_avg_m = get_mult_data()
calc_show_fstat_score(X_avg_m, y_avg_m)
```

# Verify Mult Data

```python
X_avg_s, y_avg_s = get_sub_data()
calc_show_fstat_score(X_avg_s, y_avg_s)
```
