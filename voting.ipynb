```python
import pandas as pd
from sklearn.metrics import classification_report

# 1. Load results
mlp_results = pd.read_csv('preds/mlp_kfold_results.csv')
xgb_results = pd.read_csv('preds/xgb_kfold_results.csv')

# 2. Combine and check alignment
# It is vital that the 'true_label' would have been the same for both 
# if the folds were identical.
results = pd.concat([mlp_results, xgb_results], axis=1)

# 3. Weighted Voting
# Since MLP had higher accuracy, we weigh MLP higher
w_mlp = 0.7
w_xgb = 0.3

results['ensemble_prob'] = (results['mlp_prob'] * w_mlp) + (results['xgb_prob'] * w_xgb)
results['ensemble_pred'] = (results['ensemble_prob'] > 0.5).astype(int)

# 4. Final Evaluation
print("Ensemble Performance (70 30):")
print(classification_report(results['true_label'], results['ensemble_pred']))
```
